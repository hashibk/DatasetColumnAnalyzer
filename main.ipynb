```python
import pandas as pd

df_iris = pd.read_csv('Iris.csv')
df_titanic = pd.read_csv('titanic.csv')

iris_to_find = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm']
titanic_to_find = ['Age', 'SibSp', 'Fare']
```

```python
#Data types

print('Iris: ')
for column in iris_to_find:
    print(f'{column}: {df_iris[column].dtypes}')

print()

print('Titanic: ')
for column in titanic_to_find:
    print(f'{column}: {df_titanic[column].dtypes}')
```

```python
#Total instances

print('Iris: ')
for column in iris_to_find:
    print(f'{column}: {df_iris[column].count()}')

print()

print('Titanic: ')
for column in titanic_to_find:
    print(f'{column}: {df_titanic[column].count()}')
```

```python
#Numbler of NULLS

print('Iris: ')
for column in iris_to_find:
    print(f'{column}: {df_iris[column].isnull().sum()}')

print()

print('Titanic: ')
for column in titanic_to_find:
    print(f'{column}: {df_titanic[column].isnull().sum()}')
```

```python
#Number of outliers

print ('Iris: ')
for column in iris_to_find:
    print (    column + ':')
    q1 = df_iris[column].quantile(0.25)
    q3 = df_iris[column].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr

    outliers = df_iris[(df_iris[column] < lower_bound) | (df_iris[column] > upper_bound)]
    
    # Count the number of outliers
    num_outliers = outliers[column].count()
    print(f'        Number of outliers: {num_outliers}')

print()

print ('Titanic: ')
for column in titanic_to_find:
    print (    column + ':')
    q1 = df_titanic[column].quantile(0.25)
    q3 = df_titanic[column].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr

    outliers = df_titanic[(df_titanic[column] < lower_bound) | (df_titanic[column] > upper_bound)]
    
    # Count the number of outliers
    num_outliers = outliers[column].count()
    print(f'        Number of outliers: {num_outliers}')
```

```python
# Min values
print('Iris: ')
for column in iris_to_find:
    print(f'{column}: {df_iris[column].min()}')

print()

print('Titanic: ')
for column in titanic_to_find:
    print(f'{column}: {df_titanic[column].min()}')
```

```python
# Max values
print('Iris: ')
for column in iris_to_find:
    print(f'{column}: {df_iris[column].max()}')

print()

print('Titanic: ')
for column in titanic_to_find:
    print(f'{column}: {df_titanic[column].max()}')
```

```python
# Mode values
print('Iris: ')
for column in iris_to_find:
    mode_value = df_iris[column].mode()
    print(f'{column}: {mode_value.tolist()}') 

print()

print('Titanic: ')
for column in titanic_to_find:
    mode_value = df_titanic[column].mode()
    print(f'{column}: {mode_value.tolist()}')
```

```python
# Median values
print('Iris: ')
for column in iris_to_find:
    median_value = df_iris[column].median()
    print(f'{column}: {median_value.tolist()}') 

print()

print('Titanic: ')
for column in titanic_to_find:
    median_value = df_titanic[column].median()
    print(f'{column}: {median_value.tolist()}')
```

```python
#Variance and Standard Deviation

print('Iris: ')
for column in iris_to_find:
    variance = df_iris[column].var()
    print(f'Variance of {column}: {variance}')
    std = df_iris[column].std()
    print(f'Standard deviation of {column}: {std}')

print()

print('Titanic: ')
for column in titanic_to_find:
    variance = df_titanic[column].var()
    print(f'Variance of {column}: {variance}')
    std = df_titanic[column].std()
    print(f'Standard deviation of {column}: {std}')
```

```python
import matplotlib.pyplot as plt

print('Iris: ')
for column in iris_to_find:
    print(df_iris[column].skew())
    if ((df_iris[column].skew()) > 0):
        print ('Comment: Positively Skewed')
    elif ((df_iris[column].skew()) < 0):
        print ('Comment: Negatively Skewed')
    else:
        print ('Normally distributed')

    plt.hist(df_iris[column], bins = 20, edgecolor = 'black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

    plt.boxplot(df_iris[column], vert=False)
    plt.title(f'Boxplot of {column}')
    plt.xlabel('Values')
    plt.show()

df_titanic['Age'] = df_titanic['Age'].fillna(df_titanic['Age'].median())  # Replace NaN with the median
df_titanic['Age'] = df_titanic['Age'].astype(int)  # Convert to integer

print('Titanic: ')
for column in titanic_to_find:
    print(df_titanic[column].skew())
    if ((df_titanic[column].skew()) > 0):
        print ('Comment: Positively Skewed')
    elif ((df_titanic[column].skew()) < 0):
        print ('Comment: Negatively Skewed')
    else:
        print ('Normally distributed')
        
    plt.hist(df_titanic[column], bins = 20, edgecolor = 'red')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

    plt.boxplot(df_titanic[column], vert=False)
    plt.title(f'Boxplot of {column}')
    plt.xlabel('Values')
    plt.show()
```

```python
#Missing values

print ('Iris: ')
for column in iris_to_find:
    print(f'Missing values in {column}: {df_iris[column].isnull().sum()}')

print()

print ('Titanic: ')
for column in titanic_to_find:
    print(f'Missing values in {column}: {df_titanic[column].isnull().sum()}')

#Only Age had missing values and we used to fill them with median because it
#is a numerical datatype and has outliers as well
```
